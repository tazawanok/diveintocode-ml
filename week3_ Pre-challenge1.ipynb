以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。

 ### ①ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
import matplotlib.pyplot as plt

a = len(iris_df[iris_df["Species"] == 0]) / len(iris_df["Species"])
b = len(iris_df[iris_df["Species"] == 1]) / len(iris_df["Species"])
c = len(iris_df[iris_df["Species"] == 2]) / len(iris_df["Species"])
x = np.array([a,b,c])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.pie(x, labels=['setosa', 'versicolor', 'virginica'] , autopct="%1.1f%%")
ax.set_title("pie chart")

plt.show()
```
