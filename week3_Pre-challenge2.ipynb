 ## 【問題5】用語の調査
以下の２つの単語を知る必要があります。これらを調査してまとめてください。

 - 尖度 <br>
 分布が正規分布からどれだけ歪んでいるかを表す統計量で、左右対称性を示す指標です。
「右裾が長い」もしくは「右に歪んだ」もしくは「左に偏った」分布のときには正の値を、「左裾が長い」もしくは「左に歪んだ」もしくは「右に偏った」分布のときには負の値をとります。左右対称の分布（例えば正規分布）の場合には0になります。

 <br>
 - 歪度 <br>
 分布が正規分布からどれだけ尖っているかを表す統計量で、山の尖り度と裾の広がり度を示します。
正規分布より尖った分布（データが平均付近に集中し、分布の裾が重い）のときには正の値を、正規分布より扁平な分布（データが平均付近から散らばり、分布の裾が軽い）のときには負の値をとります。正規分布の場合には0になります。
 

 ## 【問題6】分布の確認
 - seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
 - 目的変数に対して対数変換を行う。
 - 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
 - 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

 ## ①seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.distplot(housing["SalePrice"])

ax.set_xlabel('Price')
ax.set_ylabel('flequency')
ax.set_title('SalePrice')
```

```python
import scipy

print ("歪度：{}".format(scipy.stats.skew(housing["SalePrice"])))
print("尖度：{}".format(scipy.stats.kurtosis(housing["SalePrice"])))
```

 ## ②目的変数に対して対数変換を行う

```python
log_SalePrice  = np.log(housing["SalePrice"])
log_SalePrice.head()
```

 ## ③対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
fig, ax = plt.subplots()
sns.distplot(log_SalePrice)

ax.set_xlabel('Price_log')
ax.set_ylabel('flequency')
ax.set_title('SalePrice_log')
```

```python
print ("対数変換後の歪度：{}".format(scipy.stats.skew(log_SalePrice)))
print("対数変換後の尖度：{}".format(scipy.stats.kurtosis(log_SalePrice)))
```

## ④対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。
- 「SalePrice」のヒストグラムは、対数変換前は右に裾が長い。また、データが平均値付近に集中し、やや尖がった形となっている。
- 対数変換後の「SalePrice」のヒストグラムは、尖度および歪度が改善され、左右対称の正規分布に近づいている。

特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。

 - 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

 - 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

 ## ①相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
housing.corr()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(housing.corr(), annot=False)
```

 - 対角成分の相関係数は全て１になっている。
 - 「TotalBsmtSF」と「1stFlrSF」 の相関係数が、高く見える
 - 「GrLivArea」と「TotRmsAbvGrd」の相関係数が、高く見える
 - 「GarageCars」と「GarageArea」 の相関係数が、高く見える

```python
from scipy.stats import pearsonr
list_i = []
print(" <相関係数の高い特徴量の組み合わせ(0.7以上) >")
for i in housing.corr().columns:
    list_i.append(i)
    for j in housing.corr().columns:
        a, b = pearsonr(housing[i], housing[j])
        if a <1 and a > 0.7 and j not in list_i:
            print("{} and {} の相関係数 ：{}".format(i, j, a))
```

 - 相関係数が０.７を超え、強い相関のある特徴量同士の組み合わせは５つある。

 ## ②目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
dict_SalePrice = {}
for i in housing.corr().columns:
    a, b = pearsonr(housing['SalePrice'], housing[i])
    if a >= 0.5:
        dict_SalePrice[i] = a
# dict_SalePrice = sorted(dict_SalePrice.values())
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(housing[dict_SalePrice.keys()].corr(), annot=True)
```

 - 負の相関の特徴量の組み合わせはみられない。
 - 「TotalBsmtSF」と「1stFlrSF」の相関係数は０.８２となり、強い相関がある。

## ④選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
list_i = []
print("<相関係数が高い組み合わせ(3つ)>")
for i in housing[dict_SalePrice.keys()].columns:
    list_i.append(i)
    for j in housing[dict_SalePrice.keys()].columns:
        a, b = pearsonr(housing[i], housing[j])
        if a <1 and a > 0.8 and j not in list_i:
            print("{} and {} の相関係数： {}".format(i, j, a))
```
