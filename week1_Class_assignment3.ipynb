 ## 【問題】栗まんじゅうが太陽系を覆う日
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。

栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。

ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

 - 草饅頭の体積：　100立方メートル
 - 太陽系の半径：　 26500 光年(2.50709358e+20メートル)  https://ja.wikipedia.org/wiki/%E5%A4%AA%E9%99%BD%E7%B3%BB

```python
import math
manju = 100
taiyoukei_dis = 2.50709358e+20
taiyoukei_vol = (4/3) * math.pi * (taiyoukei_dis ** 3) 
minute = 0

while manju < taiyoukei_vol:
    manju *= 2
    minute += 5
else:
    print("{}分で栗まんじゅうで覆われる".format(minute))
```

```python
taiyoukei_vol
```

```python
def baibain(x=100,y=taiyoukei_vol):
    min = 0
    while x < y:
        x *= 2
        min += 5
        print("経過：{}　　総体積：{}".format(min,x))
    else:
        print("{}分でいっぱいになる".format(min))
```

```python
baibain()
```

```python
volume = 100
min = 0
v_list = [volume]
m_list = [0]
while volume < taiyoukei_vol:
    volume *=2
    v_list.append(volume)
    min += 5
    m_list.append(min)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

print("太陽系が栗まんじゅうで覆われるまでの栗まんじゅうの体積の推移")
plt.ylabel("volume", size = 14)
plt.xlabel("minute", size = 14)
plt.plot(m_list, v_list, linestyle = "--",color='blue')
plt.grid(True)
plt.show()
```
